This sheet consists of all the codes relevant to join the 9 datasets after aggregating all the datasets to the member level

```python
#importing relevant libraries

import pandas as pd
import numpy as np
from pyspark.sql.types import IntegerType, StringType, DoubleType, LongType, BooleanType, FloatType
from pyspark.sql.functions import col
#import pyspark.pandas as ps
```

```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder.master("local[1]").appName("SparkByExamples.com").getOrCreate()

# Read CSV File
df = spark.read.option("header",True).csv("humana_mays_target_members.csv")
df.printSchema()
```

```python
#dropping irrelevant columns from the dataset

df=df.drop('calendar_year')
df=df.drop('product_type')
df=df.drop('plan_category')
```

```python
#reading the datasets to be merged

df_add_feature = spark.read.option("header",True).csv("Additional Features.csv")
df_add_feature.printSchema()
```

```python
#reading the datasets to be merged

df_control_point = spark.read.option("header",True).csv("Control Point.csv")
df_control_point.printSchema()
```

```python
#merging (left join) first dataset to the additional feature dataset on id level

merged = df.join(df_add_feature, on='id', how='left')
```

```python
#merging (left join) control points dataset on id level

merged=merged.join(df_control_point, on='id', how='left')
merged.printSchema()
```

```python
#reading the datasets to be merged

df_cost_util = spark.read.option("header",True).csv("Cost & Utilization.csv")
df_cost_util.printSchema()
```

```python
#merging (left join) Cost & Utilization dataset on id level

merged=merged.join(df_cost_util, on='id', how='left')
merged.printSchema()
```

```python
#checking the shape of the merged dataset to check the number of id

print((merged.count(), len(merged.columns)))
```

```python
#reading the datasets to be merged

df_demo = spark.read.option("header",True).csv("Demographics.csv")
df_demo.printSchema()
```

```python
from pyspark.ml.feature import StringIndexer

# Initialize StringIndexer
indexer = StringIndexer(
    inputCol="rucc_category", 
    outputCol="rucc_categoryindex", 
    handleInvalid="keep"  # Strategy for handling unseen labels
)

# Fit and transform the DataFrame
df_demo = indexer.fit(df_demo).transform(df_demo)

# Clean up columns
df_demo = df_demo.drop('rucc_category')  # Remove original string column
df_demo = df_demo.withColumnRenamed('rucc_categoryindex', 'rucc_category')  # Rename index column
```

```python
#merging (left join) Demographics dataset 

merged=merged.join(df_demo, on='id', how='left')
merged.printSchema()
```

```python
#reading the datasets to be merged

df_pharm_util = spark.read.option("header",True).csv("Pharmacy Utilization.csv")
df_pharm_util.printSchema()
```

```python
#merging (left join) Pharmacy Utilization dataset 

merged=merged.join(df_pharm_util, on='id', how='left')
merged.printSchema()
```
